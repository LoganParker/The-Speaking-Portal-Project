```python
import seaborn as sns;
import pandas as pd;
import matplotlib.pyplot as plt
import math

df = pd.read_csv("AVGS.csv")
df

sns.set(rc={'figure.figsize':(14.7,8.27)})
sns.set(font_scale=1.1)
sns.set(style='dark');
b = sns.barplot(data=df, palette = "magma")
b.axes.set_title("User Feedback",fontsize=20)
b.set_xlabel("Quality Parameters",fontsize=20)
b.set_yticklabels(b.get_yticks().astype(int), size = 21)
b.set_xticklabels(df.columns, size = 11)
b.set_ylabel("Satisfaction with Parameters",fontsize=20)
plt.savefig('BarGraph.png')
```
